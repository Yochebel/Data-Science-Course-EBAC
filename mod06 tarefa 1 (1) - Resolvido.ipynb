#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("sinasc_RO_2019.csv")
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Criando gráficos de barras para visualizar contagens de APGAR1, APGAR5, SEXO, GESTACAO e ESTCIVMAE.

# Configurando tamanho do gráfico e layout
plt.rc('figure', figsize=(8, 24))
fig, axes = plt.subplots(5, 1)

# Criando gráficos de barras
sns.countplot(df, x='APGAR1', ax=axes[0])
sns.countplot(df, x='APGAR5', ax=axes[1])
sns.countplot(df, x='SEXO', ax=axes[2])
sns.countplot(df, x='GESTACAO', ax=axes[3])
sns.countplot(df, x="ESTCIVMAE", ax=axes[4])
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Visualizando a média de APGAR5 em relação às CONSULTAS por meio de um gráfico de barras.

# Configurando tamanho do gráfico
plt.rc('figure', figsize=(8, 6))
fig, axes = plt.subplots()

# Criando gráfico de barras
sns.barplot(data=df, x="APGAR5", y="CONSULTAS")
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
#Criando um painel de quatro gráficos de barras para exibir frequências de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS.

# Configurando tamanho do painel
plt.rc('figure', figsize=(12, 8))
fig, axes = plt.subplots(2, 2)

# Criando gráficos de barras para as variáveis
sns.countplot(df, x='SEXO', ax=axes[0,0])
sns.countplot(df, x='GESTACAO', ax=axes[0,1])
sns.countplot(df, x="ESTCIVMAE", ax=axes[1,0])
sns.countplot(df, x="CONSULTAS", ax=axes[1,1])

# Ajustando rótulos para melhor legibilidade
axes[0,1].set_xticklabels(axes[0,1].get_xticklabels(), rotation = 45)
plt.subplots_adjust(wspace=0.2, hspace=0.8)
```
