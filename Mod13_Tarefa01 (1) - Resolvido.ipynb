```python
import seaborn as sns
import patsy
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Ajustando um modelo de regressão linear usando a função de logaritmo da variável 'renda' como variável resposta
reg = smf.ols('''np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel +
                 qtd_filhos + tipo_renda + educacao + estado_civil + tipo_residencia +
                 idade + tempo_emprego + qt_pessoas_residencia''', data=df).fit()

# Exibindo o resumo estatístico do modelo
reg.summary()
```

```python
# Exibindo a contagem de valores únicos na coluna 'posse_de_imovel'
print(df['posse_de_imovel'].value_counts())

# Exibindo a contagem de valores únicos na coluna 'tipo_renda'
print(df['tipo_renda'].value_counts())

# Exibindo a contagem de valores únicos na coluna 'educacao'
print(df['educacao'].value_counts())

# Exibindo a contagem de valores únicos na coluna 'estado_civil'
print(df['estado_civil'].value_counts())
```

```python
# Realizando uma análise de regressão linear usando o método Ordinary Least Squares (OLS)
# A variável dependente é o logaritmo da 'renda'
# As variáveis independentes incluem características categóricas e numéricas
reg = smf.ols('''np.log(renda) ~ C(sexo, Treatment('F')) + C(posse_de_imovel, Treatment(True)) +
                 C(tipo_renda, Treatment('Assalariado')) + C(educacao, Treatment('Secundário')) +
                 C(estado_civil, Treatment('Casado')) + idade + tempo_emprego''', data=df).fit()

# Exibindo um resumo dos resultados da regressão
# O resumo inclui coeficientes, valores-p, R-quadrado e outras informações estatísticas
reg.summary()
```

```python
# Obtém os valores únicos da coluna 'tipo_renda' no DataFrame df
df['tipo_renda'].unique()
```

```python
# Mapeia os valores da coluna 'tipo_renda' para categorias desejadas
df['tipo_renda'] = df['tipo_renda'].map({
    "Assalariado": "Assalariado",
    "Pensionista": "Assalariado",
    "Bolsista": "Assalariado",
    "Empresário": "Empresário",
    "Servidor público": "Servidor público"
})

# Mapeia os valores da coluna 'educacao' para categorias desejadas
df['educacao'] = df['educacao'].map({
    "Secundário": "Fundamental",
    "Primário": "Fundamental",
    "Superior completo": "Superior",
    "Superior incompleto": "Fundamental",
    "Pós graduação": "Pós graduação"
})

# Mapeia os valores da coluna 'estado_civil' para categorias desejadas
df['estado_civil'] = df['estado_civil'].map({
    'Casado': 'Casado',
    'União': 'Não-Casado',
    'Solteiro': 'Não-Casado',
    'Separado': 'Não-Casado',
    'Viúvo': 'Não-Casado'
})

# Cria um modelo de regressão linear utilizando a biblioteca statsmodels
reg = smf.ols('''np.log(renda) ~ C(sexo, Treatment('F')) + C(posse_de_imovel, Treatment(True)) +
                 C(tipo_renda, Treatment('Assalariado')) + C(educacao, Treatment('Fundamental')) +
                 C(estado_civil, Treatment('Casado')) + idade + tempo_emprego''', data=df).fit()

# Exibe o resumo estatístico do modelo
reg.summary()
```

### Comentário:
As alterações não melhoraram muito o modelo. O R² ajustado e o AIC continuam os mesmos.

```python
# Removendo a variável 'educacao' do modelo de regressão linear
reg = smf.ols('''np.log(renda) ~ C(sexo, Treatment('F')) + C(posse_de_imovel, Treatment(True)) +
                 C(tipo_renda, Treatment('Assalariado')) + 
                 C(estado_civil, Treatment('Casado')) + idade + tempo_emprego''', data=df).fit()

# Exibe o resumo estatístico do modelo após a remoção da variável
reg.summary()
```

### Comentário: 

O R² aumentou após a remoção da variável, mas o AIC também aumentou. Isso significa que o aumento no R² não foi suficiente para compensar o aumento no AIC. Portanto, a variável removida não teve um impacto significativo no modelo, mas também não teve um impacto negativo.

```python
# Removendo a variável 'estado_civil' do modelo de regressão linear
reg = smf.ols('''np.log(renda) ~ C(sexo, Treatment('F')) + C(posse_de_imovel, Treatment(True)) +
                 C(tipo_renda, Treatment('Assalariado')) + C(educacao, Treatment('Fundamental')) +
                 idade + tempo_emprego''', data=df).fit()

# Exibe o resumo estatístico do modelo após a remoção da variável
reg.summary()
```

### Comentário:

O aumento no R² e no AIC após a remoção da variável indica uma pequena melhora no modelo.
