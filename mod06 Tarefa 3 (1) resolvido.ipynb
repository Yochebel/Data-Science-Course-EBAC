#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sinasc_RO_2019.csv")
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Definindo categorias para APGAR5
df.loc[df['APGAR5'] <= 3,'CATEGORIA'] = 'asfixia grave'
df.loc[(df['APGAR5'] >= 4) &(df['APGAR5'] <= 5),'CATEGORIA'] = 'asfixia moderada'
df.loc[(df['APGAR5'] >= 6) &(df['APGAR5'] <= 7),'CATEGORIA'] = 'asfixia leve'
df.loc[(df['APGAR5'] >= 8) &(df['APGAR5'] <= 10),'CATEGORIA'] = 'normal'

# Contagem das categorias
categoria_counts = df['CATEGORIA'].value_counts()
print(categoria_counts)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# Gráfico comparativo de APGAR5 por sexo e gestação
sns.displot(df, x='APGAR5', col='SEXO', hue='SEXO', row='GESTACAO', bins=10)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Tabela cruzada APGAR5 por tempo de gestação
freq = pd.crosstab(df['GESTACAO'], df['APGAR5'])
print(freq)
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
# Calculando os percentuais de linha
freq_percent = freq.div(freq.sum(axis=1), axis=0)
print(freq_percent)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Gráfico de barras da distribuição por percentuais de linha
freq_percent.plot.bar()
```
