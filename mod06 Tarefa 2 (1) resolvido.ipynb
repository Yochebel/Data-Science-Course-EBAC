```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando os dados
df = pd.read_csv("sinasc_RO_2019.csv")
df.columns

# Configurando estilo dos gráficos
sns.set(style="whitegrid")
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Criando um boxplot para a idade da mãe
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='IDADEMAE')
plt.title("Boxplot da Idade da Mãe")
plt.ylabel("Idade da Mãe")
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Criando um boxplot para o peso do bebê
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='PESO')
plt.title("Boxplot do Peso do Bebê")
plt.ylabel("Peso do Bebê")
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Criando um histograma para a idade da mãe
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='IDADEMAE', bins=df['IDADEMAE'].nunique()+1)
plt.title("Histograma da Idade da Mãe")
plt.xlabel("Idade da Mãe")
plt.ylabel("Contagem")
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Criando um histograma para o peso do bebê
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='PESO')
plt.title("Histograma do Peso do Bebê")
plt.xlabel("Peso do Bebê")
plt.ylabel("Contagem")
plt.show()
```
