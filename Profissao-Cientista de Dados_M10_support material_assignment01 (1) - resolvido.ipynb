Para isto, conduziu um estudo com alguns clientes, comprovando suas rendas através de olerites e outros documentos, e pretende construir um modelo preditivo para esta renda com base em algumas variáveis que já possui em seu banco de dados.

As variáveis são intuitivas - note que há uma variável 'index' que é um identificador do cliente e que em geral o ```read_csv``` do pandas coloca também uma variável sequencial.

Estes dados estão no arquivo ```previsao_de_renda.csv```. Carregue-o em um *dataframe*.

1) Avalie a estrutura de correlação das variáveis quantitativas através de um gráfico de "matriz de dispersão" e através da avaliação gráfica da matriz de correlações. Veja se você identifica algum padrão interessante ou que te faça sentido.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Importa a biblioteca pandas e lê um arquivo CSV chamado "previsao_de_renda.csv"
df = pd.read_csv("previsao_de_renda.csv", index_col='index')

# Remove a coluna chamada "Unnamed: 0" do DataFrame 'df' ao longo do eixo das colunas (axis=1)
df = df.drop("Unnamed: 0", axis=1)

# O DataFrame 'df' agora contém os dados do arquivo CSV com o índice configurado para a coluna 'index'
df
````

````python
# Lista de variáveis quantitativas
quantitativas = ["qtd_filhos", "idade", "tempo_emprego", "qt_pessoas_residencia", "renda"]

# Calcula a matriz de correlação entre as variáveis quantitativas no DataFrame 'df'
correlacao = df[quantitativas].corr()

# O DataFrame 'correlacao' agora contém os coeficientes de correlação entre as variáveis quantitativas
df[quantitativas].corr()
````

````python
# Define uma paleta de cores divergente usando a função diverging_palette do seaborn
# A paleta é usada para a representação visual da matriz de correlação
cmap = sns.diverging_palette(h_neg=215, h_pos=355, s=100, l=50, as_cmap=True, sep=40, center='light')

# Cria um clustermap utilizando a matriz de correlação entre as variáveis quantitativas no DataFrame 'df'
# figsize define o tamanho da figura
# center define o ponto central da escala de cores
# cmap especifica a paleta de cores a ser utilizada
sns.clustermap(df[quantitativas].corr(), figsize=(8, 8), center=0, cmap=cmap)
````

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
# Adiciona uma nova coluna chamada 'renda_log' ao DataFrame 'df'
# A coluna 'renda_log' contém o logaritmo natural (base e) dos valores da coluna 'renda'
df['renda_log'] = np.log(df['renda'])

# O DataFrame 'df' agora inclui a coluna 'renda_log' com os valores transformados pelo logaritmo natural
df
````

````python
# Cria um gráfico de dispersão (scatterplot) utilizando a biblioteca seaborn
# Os dados são representados no eixo x pela posição do índice do DataFrame 'df' e no eixo y pelos valores da coluna 'renda_log'
sns.scatterplot(data=df, x=df.index, y='renda_log')
````

````python
# Cria um boxplot utilizando a biblioteca seaborn
# Os dados são representados no eixo y pelos valores da coluna 'renda_log' do DataFrame 'df'
sns.boxplot(data=df, y='renda_log')
````

5) A aplicação do LOG você acha que melhorou ou piorou a análise?

Ao realizar a análise por meio do gráfico de dispersão, percebi uma dificuldade em interpretar as tendências ou padrões nos dados. No entanto, ao optar pela abordagem do boxplot, a visualização tornou-se consideravelmente mais clara e legível. Ja que o boxplot oferece uma representação mais clara das estatísticas descritivas, facilitando a identificação de medidas centrais, dispersão e possíveis valores atípicos na distribuição dos dados.
